```python
import pandas as pd
import matplotlib.pyplot as plt

log_data = pd.read_csv('train_log.csv')

print(log_data)
# Plot training & validation Mean Average Error for the bounding box
plt.figure(figsize=(12, 10))
plt.subplot(2, 2, 1)
plt.plot(log_data['epoch'], log_data['bbox_reshaped_mae'], label='Train')
plt.plot(log_data['epoch'], log_data['val_bbox_reshaped_mae'], label='Validation')
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='upper left')

# Plot training & validation loss values
plt.subplot(2, 2, 2)
plt.plot(log_data['epoch'], log_data['loss'], label='Train')
plt.plot(log_data['epoch'], log_data['val_loss'], label='Validation')
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper left')

# Plot training & validation accuracy for the labels 
plt.subplot(2, 2, 3)
plt.plot(log_data['epoch'], log_data['label_reshaped_binary_accuracy'], label='Train')
plt.plot(log_data['epoch'], log_data['val_label_reshaped_binary_accuracy'], label='Validation')
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper left')

plt.show()
```
